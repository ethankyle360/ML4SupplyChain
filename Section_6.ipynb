### What is the most popular destination city from New York?

```python
flights.columns
table1 = flights.groupby('dest').agg(count=('dest','count')).sort_values(by='count',ascending=False).reset_index()
```

```python
pd.merge(table1, airports[['faa', 'name']], how='left', left_on='dest', right_on='faa')
```

Conclusion: Chicago is the most popular destination from New York.

### Which month is the busiest of the year?

```python
flights.month.value_counts()
```

Conclusion: The month of July is the busiest of the year.

### Which airline is the most punctual?

```python
flights['total_delay'] = flights['arr_delay'] + flights['dep_delay']

table1 = flights.groupby('carrier').agg(mean_delay=('total_delay', 'mean')).sort_values(by='mean_delay').reset_index()
```

```python
airlines
pd.merge(table1, airlines, how='left')
```

Conclusion: Hawaiian Airlines is the most punctual since it has the closest value to 0. Also, Frontier Airlines is the worst in terms of delays.

### Which destination has the longest duration?

```python
flights.air_time.describe()
```

```python
table1 = flights.groupby(['origin','dest']).agg(average_air_time=
 ('air_time', 'mean')).reset_index().sort_values(by='average_air_time', ascending=False)
```

```python
pd.merge(table1, airports[['faa', 'name']], how='left', left_on='origin', right_on='faa')
```

Conclusion: Hononlulu is the longest duration destination.

### Which airline has the highest capacity of seats?

```python
planes.columns
flights.columns

carrier_tailnum = flights[['carrier', 'tailnum']].drop_duplicates()
```

```python
seats = pd.merge(carrier_tailnum, planes[['tailnum', 'seats']], how='left').groupby('carrier').agg(total_seats=('seats', 'sum')).sort_values(by='total_seats', ascending=False)
```

```python
seats
```

Conclusion: United Airlines has the highest seat capacity.

### Which airplane model is the highest in use and from the manufacturer?

```python
airplanes_use = flights.groupby('tailnum').agg(count= ('tailnum', 'count')).reset_index()

pd.merge(planes[['tailnum', 'model', 'manufacturer']], airplanes_use).groupby(['model', 'manufacturer']).agg(total_flights= ('count', 'sum')).sort_values(by='total_flights', ascending=False) # Use a valid tuple to define the aggregation
```

Conclusion: Airbus A320-232 is the highest model in use.
